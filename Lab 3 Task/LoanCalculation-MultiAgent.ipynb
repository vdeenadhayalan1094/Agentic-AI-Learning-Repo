```python
def calculate_loan(principal, rate, time):
    """
    Simple Interest Formula:
    SI = (P * R * T) / 100
    Total = P + SI
    """
    si = (principal * rate * time) / 100
    total = principal + si
    return {
        "simple_interest": si,
        "total_amount": total
    }
```

```python
def agent_1_extract(user_query):
    prompt = f"""
Extract the following details from the user query:
- loan_amount
- interest_rate
- duration (in years)

Return strictly in JSON format:
{{
  "loan_amount": number,
  "interest_rate": number,
  "duration": number
}}

Query: {user_query}
"""
    
  response = client.chat.completions.create(
    model="llama3-8b-8192",   # ✅ UPDATED
    messages=[{"role": "user", "content": prompt}],
    temperature=0
)
    
    output = response.choices[0].message.content
    print("Agent 1 Output:\n", output, "\n")
    
    return eval(output)  # simple parsing
```

```python
def agent_2_calculate(data):
    principal = data["loan_amount"]
    rate = data["interest_rate"]
    time = data["duration"]
    
    result = calculate_loan(principal, rate, time)
    
    print("Agent 2 Calculation:\n", result, "\n")
    
    return {
        **data,
        **result
    }
```

```python
def agent_3_verify(data):
    prompt = f"""
Verify the following loan calculation:

Principal: {data['loan_amount']}
Rate: {data['interest_rate']}
Time: {data['duration']}
Calculated Interest: {data['simple_interest']}
Total Amount: {data['total_amount']}

If correct, confirm and present final answer clearly.
"""
    
    response = client.chat.completions.create(
    model="llama3-8b-8192",   # ✅ UPDATED
    messages=[{"role": "user", "content": prompt}],
    temperature=0
)
    
    output = response.choices[0].message.content
    print("Agent 3 Verification:\n", output)
```

```python
def run_multi_agent(user_query):
    print("User Query:\n", user_query, "\n")
    
    # Agent 1
    extracted_data = agent_1_extract(user_query)
    
    # Agent 2
    calculated_data = agent_2_calculate(extracted_data)
    
    # Agent 3
    agent_3_verify(calculated_data)
```
